```python
hmovies11=hmovies1.drop(['LeadStudio','Story','Genre'],axis=1)
```

```python
hmovies12=hmovies1[['LeadStudio','Story','Genre']]
```

```python
#Filling missing values to numerical variables
from sklearn.impute import SimpleImputer
mp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
mp_median = SimpleImputer(missing_values=np.nan, strategy='median')
hmovies11=mp_median.fit_transform(hmovies11)
```

```python
hmovies11
```

```python
#filling missing values for categorical data
hmovies12=hmovies12.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
hmovies12.isna().sum()
```

```python
hmovies13=pd.DataFrame(hmovies11)
```

```python
hmovies13.columns=n
```

```python
hmovies=pd.concat([hmovies2,hmovies12,hmovies13],axis=1)
```

```python
hmovies
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
```

```python
hmovies.dtypes
```

```python
hmovies['Movie']=lb_make.fit_transform(hmovies['Movie'])
hmovies['LeadStudio']=lb_make.fit_transform(hmovies['LeadStudio'])
hmovies['Story']=hmovies['Story'].astype('category')
hmovies['Genre']=hmovies['Genre'].astype('category')
hmovies['Story']=hmovies['Story'].cat.codes
hmovies['Genre']=hmovies['Genre'].cat.codes
```

```python
hmovies.head()
```

```python
from scipy import stats
```

```python
#Outlier Analysis
z=np.abs(stats.zscore(hmovies))
#np.where(z>3)
hmovies = hmovies[(z < 3).all(axis=1)]
```
